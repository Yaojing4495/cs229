```python
import time
from sklearn.metrics import log_loss
from sklearn.model_selection import StratifiedKFold
```

```python
N_SPLITS = 5
strat_kf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=50)

scores = np.empty(N_SPLITS)
for idx, (train_idx, test_idx) in enumerate(strat_kf.split(train_X,train_y)):
    print("=" * 12 + f"Training fold {idx}" + 12 * "=")
    start = time.time()

    X_train, X_val = train_X.iloc[train_idx], train_X.iloc[test_idx]
    y_train, y_val = train_y[train_idx], train_y[test_idx]
    eval_set = [(X_val, y_val)]

    lgbm_clf = lgb.LGBMClassifier(n_estimators=10000)
    lgbm_clf.fit(
        X_train,
        y_train,
        eval_set=eval_set,
        early_stopping_rounds=200,
        eval_metric="binary_logloss",
        verbose=False,
    )

    preds = lgbm_clf.predict_proba(X_val)
    loss = log_loss(y_val, preds)
    scores[idx] = loss
    runtime = time.time() - start
    print(f"Fold {idx} finished with score: {loss:.5f} in {runtime:.2f} seconds.\n")
```

```python
## for LGBMRanker

N_SPLITS = 5
strat_kf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=50)

scores = np.empty(N_SPLITS)
for idx, (train_idx, test_idx) in enumerate(strat_kf.split(train_X,train_y)):
    print("=" * 12 + f"Training fold {idx}" + 12 * "=")
    start = time.time()

    X_train, X_val = train_X.iloc[train_idx], train_X.iloc[test_idx]
    y_train, y_val = train_y[train_idx], train_y[test_idx]
    eval_set = [(X_val, y_val)]
    
    #test_baskets = X_val.groupby(['week', 'customer_id'])['article_id'].count().values
    #train_baskets = X_train.groupby(['week', 'customer_id'])['article_id'].count().values
    ranker = LGBMRanker(
        objective="lambdarank",
        metric="logloss",
        boosting_type="dart",
        n_estimators=3,
        importance_type='gain',
        verbose=10,
    )
    
    ranker.fit(
        X_train,
        y_train,
        group=train_baskets,
        eval_set=eval_set,
        early_stopping_rounds=200,
        eval_metric="binary_logloss",
        verbose=False,
        eval_group=[test_baskets]
    )

    preds = ranker.predict_proba(X_val)
    loss = log_loss(y_val, preds)
    scores[idx] = loss
    runtime = time.time() - start
    print(f"Fold {idx} finished with score: {loss:.5f} in {runtime:.2f} seconds.\n")
   
```

#### First, create a CV splitter — we are choosing StratifiedKFold because it is a classification problem. Then, loop through each train/test sets using split. In each fold, initialize and train a new LGBM model and optionally report the score and runtime. 

boosting_type (str, optional (default='gbdt')) – ‘gbdt’, traditional Gradient Boosting Decision Tree. ‘dart’, Dropouts meet Multiple Additive Regression Trees. ‘goss’, Gradient-based One-Side Sampling. ‘rf’, Random Forest.
