```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python

mae1 = pd.read_excel("mae.xlsx")
rmse1 = pd.read_excel("rmse.xlsx")

mae = pd.read_excel("mae1.xlsx")
rmse = pd.read_excel("rmse1.xlsx")
```

```python
plt.figure().set_size_inches(100,100)
plt.figure(12)

plt.subplot(121)
plt.plot(mae1["iterations"],mae1["K = 2"], label = ' K = 2, GD', linewidth = 1, color = 'r')
plt.plot(mae1["iterations"],mae1["K = 3"], label = ' K = 3, GD', linewidth = 1, color = 'g')
plt.plot(mae1["iterations"],mae1["K = 4"], label = ' K = 4, GD', linewidth = 1, color = 'b')
plt.plot(mae1["iterations"],mae1["K = 5"], label = ' K = 5, GD', linewidth = 1, color = 'y')

plt.plot(mae["iterations"],mae["K = 2"], label = ' K = 2, SVD', linewidth = 1, color = 'c')
plt.plot(mae["iterations"],mae["K = 3"], label = ' K = 3, SVD', linewidth = 1, color = 'm')
plt.plot(mae["iterations"],mae["K = 4"], label = ' K = 4, SVD', linewidth = 1, color = 'tan')
plt.plot(mae["iterations"],mae["K = 5"], label = ' K = 5, SVD', linewidth = 1, color = 'k')
plt.xlabel("iterations")
plt.ylabel('error')
plt.title("MAE")
plt.legend()
plt.grid()

plt.subplot(122)
plt.plot(rmse1["iterations"],rmse1["K = 2"], label = ' K = 2, GD', linewidth = 1, color = 'r')
plt.plot(rmse1["iterations"],rmse1["K = 3"], label = ' K = 3, GD', linewidth = 1, color = 'g')
plt.plot(rmse1["iterations"],rmse1["K = 4"], label = ' K = 4, GD', linewidth = 1, color = 'b')
plt.plot(rmse1["iterations"],rmse1["K = 5"], label = ' K = 5, GD', linewidth = 1, color = 'y')

plt.plot(rmse["iterations"],rmse["K = 2"], label = ' K = 2, SVD', linewidth = 1, color = 'c')
plt.plot(rmse["iterations"],rmse["K = 3"], label = ' K = 3, SVD', linewidth = 1, color = 'm')
plt.plot(rmse["iterations"],rmse["K = 4"], label = ' K = 4, SVD', linewidth = 1, color = 'tan')
plt.plot(rmse["iterations"],rmse["K = 5"], label = ' K = 5, SVD', linewidth = 1, color = 'k')

plt.xlabel("iterations")
plt.title("RMSE")
plt.legend()
plt.grid()
fig = plt.gcf()
fig.set_size_inches(20,5)
plt.show
plt.savefig('error')
```
