```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
mse = pd.read_excel("mse1.xlsx")
mae = pd.read_excel("mae1.xlsx")
rmse = pd.read_excel("rmse1.xlsx")
```

```python
plt.figure().set_size_inches(100,100)
plt.figure(13)
plt.subplot(131)
plt.plot(mse["iterations"],mse["K = 2"], label = ' K = 2', linewidth = 1, color = 'r')
plt.plot(mse["iterations"],mse["K = 3"], label = ' K = 3', linewidth = 1, color = 'g')
plt.plot(mse["iterations"],mse["K = 4"], label = ' K = 4', linewidth = 1, color = 'b')
plt.plot(mse["iterations"],mse["K = 5"], label = ' K = 5', linewidth = 1, color = 'y')
plt.xlabel("iterations")
plt.ylabel('error')
plt.title("mse")
plt.legend()
plt.grid()

plt.subplot(132)
plt.plot(mae["iterations"],mae["K = 2"], label = ' K = 2', linewidth = 1, color = 'r')
plt.plot(mae["iterations"],mae["K = 3"], label = ' K = 3', linewidth = 1, color = 'g')
plt.plot(mae["iterations"],mae["K = 4"], label = ' K = 4', linewidth = 1, color = 'b')
plt.plot(mae["iterations"],mae["K = 5"], label = ' K = 5', linewidth = 1, color = 'y')
plt.xlabel("iterations")
#plt.ylabel('error')
plt.title("mae")
plt.legend()
plt.grid()

plt.subplot(133)
plt.plot(rmse["iterations"],rmse["K = 2"], label = ' K = 2', linewidth = 1, color = 'r')
plt.plot(rmse["iterations"],rmse["K = 3"], label = ' K = 3', linewidth = 1, color = 'g')
plt.plot(rmse["iterations"],rmse["K = 4"], label = ' K = 4', linewidth = 1, color = 'b')
plt.plot(rmse["iterations"],rmse["K = 5"], label = ' K = 5', linewidth = 1, color = 'y')
plt.xlabel("iterations")
#plt.ylabel('error')
plt.title("rmse")
plt.legend()
plt.grid()
fig = plt.gcf()
fig.set_size_inches(12,3)
plt.show
plt.savefig('error')
```
