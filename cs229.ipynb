```python
customer_ids = customer_sample['customer_id']
customer_ids_list = customer_ids.values.tolist()
len(customer_ids_list)
```

```python
article_ids = transactions_with_article_customers_info['article_id']

article_ids_list = []
[article_ids_list.append(x) for x in article_ids if x not in article_ids_list]

len(article_ids_list)
```

```python
customer_article_matrix = np.zeros(shape=(len(customer_ids_list),len(article_ids_list)))
```

```python
customer_article_matrix = pd.DataFrame(columns=customer_ids_list, index=article_ids_list)
```

```python
customer_article_matrix = customer_article_matrix.fillna(0.0)
customer_article_matrix
```

```python
for i in range(len(transactions_with_article_customers_info)):
    customer = transactions_with_article_customers_info.iloc[i]['customer_id']
    article = transactions_with_article_customers_info.iloc[i]['article_id']
    customer_article_matrix.loc[article,customer] = 1
```
